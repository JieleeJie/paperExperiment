```python
import numpy as np

# 任务数量
task_number = 90

# 任务大小 服从 均值=4600M条指令 的 均匀分布
xiaoshu, task_size = np.modf(np.random.uniform(2000,7200,task_number))
# 移动设备处理能力(MIPS)
speed_moblieDevice = 500
# manet处理能力(MIPS)
speed_manet = 1200
# cloudlet处理能力(MIPS)
speed_cloudlet = 3846
# clooud处理能力(MIPS)
speed_cloud = 31790

# 任务所需传输数据量 服从 均值=750KB 的 均匀分布
# 假设下行的时间为上行的 0.6 倍
xiaoshu1, task_upload = np.modf(np.random.uniform(200,1300,task_number))
# 边缘传输速率：带宽设为100mbps 理论传输率最大 12.5MB/s  实际数据传输率 4-11MB/S
xiaoshu2,trans_rate_clt = np.modf(np.random.uniform(4,11,task_number))
# manet传输速率：带宽设为10mbps 理论传输率最大 1.25MB/s  实际数据传输率 400KB/S-1MB/S
trans_rate_manet =np.random.uniform(0.4,1,task_number)
# 云传输速率：带宽设为1Gbps 理论传输率最大 128MB/s  实际数据传输率 40-110MB/S
xiaoshu4,trans_rate_c = np.modf(np.random.uniform(40,110,task_number))

# 功率
# 设 移动设备空闲时的功率为 5mW/sec 通信时的能耗为 20mW/sec   1w = 1 J/S     1mw = 1e-3 J/S
power_idle = 5
power_communication = 20
power_excute = 60

totalTime = 0
totalEnergy = 0
```

```python
# local 1%的任务量   cloudlet 75%的任务量 cloud 24%的任务量
# 总的时间 和 能耗
Tm = Tmanet = Tclt = Tc = 0
Em = Emanet =  Eclt = Ec = 0
temp = 0
Tmanet_upload = Tmanet_queue = Tmanet_excute = 0
Tclt_upload = Tclt_queue = Tclt_excute = 0
Tc_upload = Tc_excute = 0
# 任务失败个数
failureCount = 0


for i in range(task_number):
    if i <= task_number * 0.75:
        Tclt_upload = (task_upload[i] / 1024) / trans_rate_clt[i]
        Tclt_queue = np.random.uniform(0,1.5)
        Tclt_excute = task_size[i] / speed_cloudlet
        if Tclt_upload * 1.6  + Tclt_queue + Tclt_excute > 10:
            failureCount += 1
        Tclt = Tclt + Tclt_upload * 1.6  + Tclt_queue + Tclt_excute
        Eclt = Eclt +  power_communication * Tclt_upload * 1.6 + power_idle * ( Tclt_queue + Tclt_excute)
    else:
        Tc_upload = (task_upload[i] / 1024) / trans_rate_c[i]
        Tc_excute = task_size[i] / speed_cloud
        if Tc_upload * 1.6 + Tc_excute > 10:
            failureCount += 1
        Tc = Tc + Tc_upload * 1.6 + Tc_excute
        Ec = Ec + power_communication * Tc_upload * 1.6 + power_idle * Tc_excute
totalTime += Tclt + Tc
totalEnergy += Eclt + Ec
print("任务比例1的时间约为：",round(Tclt + Tc,2))
print("任务比例1的能耗约为：", round(Eclt + Ec,2))
print("任务比例1的失败个数为：",failureCount)
print('----------------------------------------')

# local 16%的任务量   manet 84%的任务量
# 总的时间 和 能耗
Tm = Tmanet = Tclt = Tc = 0
Em = Emanet =  Eclt = Ec = 0
temp = 0
Tmanet_upload = Tmanet_queue = Tmanet_excute = 0
Tclt_upload = Tclt_queue = Tclt_excute = 0
Tc_upload = Tc_excute = 0
# 任务失败个数
failureCount = 0

for i in range(task_number):
    if i <= task_number*0.16:
        temp = task_size[i] / speed_moblieDevice
        if temp > 10:
            failureCount += 1
        Tm += temp
        Em += power_excute * temp
    else:
        Tmanet_upload = (task_upload[i] / 1024) / trans_rate_manet[i]
        Tmanet_queue = np.random.uniform(0,1.5)
        Tmanet_excute = task_size[i] / speed_manet
        if Tmanet_upload * 1.6  + Tmanet_queue + Tmanet_excute > 10:
            failureCount += 1
        Tmanet = Tmanet + Tmanet_upload * 1.6  + Tmanet_queue + Tmanet_excute
        Emanet = Emanet +  power_communication * Tmanet_upload * 1.6 + power_idle * ( Tmanet_queue + Tmanet_excute)
totalTime += Tm + Tmanet
totalEnergy += Em + Emanet
print("任务比例2的时间约为：",round(Tm + Tmanet ,2))
print("任务比例2的能耗约为：", round(Em + Emanet ,2))
print("任务比例2的失败个数为：",failureCount)
print('----------------------------------------')


# local 10%的任务量 cloudlet 70%的任务量 cloud 20%的任务量
# 总的时间 和 能耗
Tm = Tmanet = Tclt = Tc = 0
Em = Emanet =  Eclt = Ec = 0
temp = 0
Tmanet_upload = Tmanet_queue = Tmanet_excute = 0
Tclt_upload = Tclt_queue = Tclt_excute = 0
Tc_upload = Tc_excute = 0
# 任务失败个数
failureCount = 0

for i in range(task_number):
    if i <= task_number*0.1:
        temp = task_size[i] / speed_moblieDevice
        if temp > 10:
            failureCount += 1
        Tm += temp
        Em += power_excute * temp
    elif task_number*0.1 < i < task_number * 0.8:
        Tclt_upload = (task_upload[i] / 1024) / trans_rate_clt[i]
        Tclt_queue = np.random.uniform(0,1.5)
        Tclt_excute = task_size[i] / speed_cloudlet
        if Tclt_upload * 1.6  + Tclt_queue + Tclt_excute > 10:
            failureCount += 1
        Tclt = Tclt + Tclt_upload * 1.6  + Tclt_queue + Tclt_excute
        Eclt = Eclt +  power_communication * Tclt_upload * 1.6 + power_idle * ( Tclt_queue + Tclt_excute)
    else:
        Tc_upload = (task_upload[i] / 1024) / trans_rate_c[i]
        Tc_excute = task_size[i] / speed_cloud
        if Tc_upload * 1.6   + Tc_excute > 10:
            failureCount += 1
        Tc = Tc +  Tc_upload * 1.6   + Tc_excute
        Ec = Ec + power_communication * Tc_upload * 1.6 + power_idle *  Tc_excute
totalTime += Tm + Tclt + Tc
totalEnergy += Em + Eclt + Ec
print("任务比例3的时间约为：",round(Tm + Tclt + Tc ,2))
print("任务比例3的能耗约为：", round(Em + Eclt + Ec ,2))
print("任务比例3的失败个数为：",failureCount)
print('----------------------------------------')
```
